```python
import pickle

dataset='DataSet1'

master_all_df = pickle.load(open(f'{dataset}/real_master_df2.dump', 'rb'))
vdc_capacity_dic = pickle.load(open(f'{dataset}/vdc_capacity_dic.dump', 'rb'))
VDC = list(vdc_capacity_dic.keys())

VDC_without_3A = VDC.copy()
VDC_without_3A.remove("3A")
```

```python
df_without_3A = master_all_df[~master_all_df['vin'].isin(list(master_all_df[master_all_df["vdc"]=='3A']['vin'].unique()))] 
```

```python
# vdc = "BE"
# df_without_3A[df_without_3A["next"]==vdc]

import pandas as pd
from datetime import timedelta
st = pd.to_datetime('2011-01-01')
arrival_time_list=[]

while st != pd.to_datetime('2012-12-31 23:00'):
    
    arrival_time_list.append(st)
    st+=timedelta(hours=1)
```

```python
vdc_inventory_dic={}
vdc_current_inventory_dic={}
for n_vdc in VDC_without_3A:
    print(n_vdc)
    
    start_inventory = 0
    len_inventory_list=[0]
    asdf= [0]
    
    df = df_without_3A[df_without_3A["vdc"]==n_vdc]
    
    # print(df)
    for index, time in enumerate(arrival_time_list[1:]):
        # print(df[df["arrival_time"]<time & (df["arrival_time"]>=arrival_time_list[index-1])])
        t_in = len(df[(df["arrival_time"]<time) & (df["arrival_time"]>=arrival_time_list[index])].index)
        #print(t_in)
        len_inventory_list.append(t_in - start_inventory)
        asdf.append(t_in)
        start_inventory = t_in
    
    #print(n_vdc)    
    #print(len_inventory_list)
    #print(asdf)
    
    print(len_inventory_list)
    
    vdc_inventory_dic[n_vdc] = len_inventory_list
    vdc_current_inventory_dic[n_vdc] = asdf
```
